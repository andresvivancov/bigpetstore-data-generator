```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
from collections import defaultdict
```

```python
def read_transactions(flname):
    fl = open(flname)
    transactions = list()
    trans_cust_id = None
    products = None
    for ln in fl:
        trans_id, trans_time, store_id, store_zipcode, store_city, store_state, customer_id, customer_name, customer_zipcode, customer_city, customer_state, product = ln.split(",")
        if trans_cust_id != (trans_id, customer_id):
            if trans_cust_id != None:
                transaction = (int(trans_id), float(trans_time), int(store_id), store_zipcode, store_city, store_state,
                         int(customer_id), customer_name, customer_zipcode, customer_city, customer_state, products)
                transactions.append(transaction)
            products = []
            trans_cust_id = (trans_id, customer_id)
        products.append(product.strip())
    fl.close()
    return transactions
```

```python
no_weather_transactions = read_transactions("../generatedDataNoWeather/transactions.txt")
```

```python
stores_customers = defaultdict(set)
for trans in no_weather_transactions:
    store = str(trans[2]) + "_" + trans[4] + "_" + trans[5]
    cust = str(trans[6]) + "_" + trans[9] + "_" + trans[10]
    stores_customers[store].add(cust)
```

```python
all_stores = stores_customers.keys()
customer_counts = [len(stores_customers[store]) for store in all_stores]

plt.bar(np.arange(len(all_stores)), customer_counts, 0.5)
plt.xlabel("Store", fontsize=16)
plt.ylabel("Customers", fontsize=16)
```

```python
def read_weather(flname):
    fl = open(flname)
    all_weather = defaultdict(list)
    for ln in fl:
        cols = ln.strip().split(",")
        time = float(cols[0])
        store = int(cols[1])
        temp = float(cols[2])
        windchill = float(cols[3])
        windspeed = float(cols[4])
        precip = float(cols[5])
        rainfall = float(cols[6])
        snowfall = float(cols[7])
        record = (time, temp, windchill, windspeed, precip, rainfall, snowfall)
        all_weather[store].append(record)
    fl.close()
    return all_weather
```

```python
weather_data = read_weather("../generatedDataWeather/weather.txt")
```

```python
plt.hold(True)
for store, weather in weather_data.iteritems():
    weather = np.array(weather)
    plt.plot(weather[:, 0], weather[:, 1], label=store)
    #plt.hist(weather[:, 1], bins=50)
plt.legend()
```
